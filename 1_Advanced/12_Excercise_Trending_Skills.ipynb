```python
import ast 
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('D:\Projects\Python_Data_Project\data_jobs.csv')

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
def clean_list(skills_list):
    if pd.notna(skills_list):
        return ast.literal_eval(skills_list)

df['job_skills'] = df['job_skills'].apply(clean_list)
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()

df_DA
```

```python
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month

df_DA_exploded = df_DA.explode('job_skills')
```

```python
df_DA_pivot = df_DA_exploded.pivot_table(index = 'job_posted_month_no', columns='job_skills', aggfunc = 'size', fill_value=0) #fill value removes/fill NANS with 0s or given value
```

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_pivot = df_DA_pivot.drop('Total')
```

```python
df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5].plot(kind = 'line')
plt.title('Top 5 skills for Data Analysts per Month')
plt.xlabel("Month Number")
plt.ylabel('Job count')
plt.show()
```

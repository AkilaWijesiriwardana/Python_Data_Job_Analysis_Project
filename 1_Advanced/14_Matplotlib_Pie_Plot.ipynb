```python
import ast
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('D:\Projects\Python_Data_Project\data_jobs.csv')

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
def clean_list(skills_list):
    if pd.notna(skills_list):
        return ast.literal_eval(skills_list)

df['job_skills'] = df['job_skills'].apply(clean_list)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle = 90, autopct='%1.1f%%')
plt.title('Work from Home Status')
plt.xlabel('')
plt.show()
```

```python
fig, ax = plt.subplots(1,3)

df['job_health_insurance'].value_counts().plot(kind='pie', ax=ax[0], title='Insurance', startangle=90, autopct ='%1.1f%%')
df['job_work_from_home'].value_counts().plot(kind='pie', ax=ax[1], title='Work from home', startangle=90, autopct = "%1.1f%%")
df['job_no_degree_mention'].value_counts().plot(kind='pie', ax=ax[2], title ='Degree mentioned?',startangle=90,autopct ='%1.1f%%')
fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1,3)

dict_column = {
    'job_work_from_home':'Work From Home',
    'job_no_degree_mention':'Degree req.',
    'job_health_insurance': 'Health Insurance'

}
for i,(column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle=90, autopct = '%1.1f%%', labels =['False','True'])
    ax[i].set_title(title)

plt.show()
```
